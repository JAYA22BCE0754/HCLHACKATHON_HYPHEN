# Regression Hackathon Workflow

This notebook demonstrates loading data, running EDA, training multiple regression models, and exporting artifacts for API inference.

```python
from pathlib import Path
import json
import pandas as pd

project_root = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
artifacts = project_root / 'artifacts'
artifacts.mkdir(parents=True, exist_ok=True)
print('Project root:', project_root)
```

```python
from src.data import create_sample_dataset, load_dataset, validate_schema

data_path = project_root / 'data' / 'raw' / 'housing_sample.csv'
if not data_path.exists():
    create_sample_dataset(data_path)

df = load_dataset(data_path)
validate_schema(df, 'SalePrice')
df.head()
```

```python
import subprocess

cmd = ['python', '-m', 'src.train', '--data-path', 'data/raw/housing_sample.csv', '--target', 'SalePrice']
result = subprocess.run(cmd, capture_output=True, text=True)
print(result.stdout)
if result.returncode != 0:
    print(result.stderr)
```
